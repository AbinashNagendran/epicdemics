```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import os 


print(os.listdir())
# Assign directory
directory = r"./D-FENSE/processed_data"
findf = pd.DataFrame()
# Iterate over files in directory
for name in os.listdir(directory):
    if name[-3:] == "csv":
        citycode = name[-6:-4]
        df = pd.read_csv(os.path.join(directory, name))
        df['region'] = citycode
        if len(findf) == 0:
            findf = df
        else:
            findf = pd.concat([findf, df])
        print("-----")
        #print(f.read())
findf = findf.sort_values('epiweek')
print(findf)


#df = pd.read_csv("D-FENSE/processed_data/.csv")
```

```python
findf.to_csv('merged_d-fense.csv', index=False)
```
